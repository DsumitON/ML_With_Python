```python
#Problem Statement: Given the data set of adv predict sales based on TV ads
```

```python
#Import libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import metrics  #For evaluation o the model
```

```python
#Read Data from Data set
data=pd.read_csv('Advertising.csv')
print(data.head())
```

```python
#Check linearity
plt.scatter(data['TV'],data['sales'])
plt.xlabel("TV Add Exp")
plt.ylabel("Sales")
plt.show()
```

```python
#Data Preprocessing
#Check null values
data.isnull().sum()
```

```python
#Segrete Ind and Dep variable
x=data['TV'].values.reshape(-1,1)  #Independent Variable
y=data['sales'].values.reshape(-1,1)  #Dependent Variable
print(x)
```

```python
#Traing the Linear Regression Algo
lin_reg=LinearRegression()  #Creating object of Algo
lin_reg.fit(x,y)
```

```python
#We got trained model
```

```python
#Lets do some predictions
y_pred=lin_reg.predict(x)
print(y_pred)
```

```python
#Showing Reg Line
plt.scatter(data['TV'],data['sales'])
plt.plot(data['TV'],y_pred,color='red')
plt.xlabel("TV Add Exp")
plt.ylabel("Sales")
plt.show()
```

```python
#Coeficients
print(lin_reg.coef_) #b1
print(lin_reg.intercept_) #b0
```

```python
#Prediction on unknow Values
TV=1050
pred_sales=lin_reg.predict([[TV]])
print(pred_sales)
```

```python
print("Mean squared Error(MSE):",metrics.mean_squared_error(y_pred,y))
print("Root Mean Squared Erro(RMSE):",np.sqrt(metrics.mean_squared_error(y_pred,y)))
print("Accuracy:",100-metrics.mean_squared_error(y_pred,y))
```
