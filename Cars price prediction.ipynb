Context

Riding the digital wave, India's used car market is set to grow at a compounded annual growth of 11% and is likely to touch sales of up to 8.3 million units by FY26 as more people have been opting for pre-owned cars for personal mobility in the pandemic amid the ongoing supply shortages for manufacturing new cars.

The used car market in the country is expected to reach over 70 lakh vehicles by 2025-26, up from 38 lakh in 2020-21 as the Covid-19 pandemic, digitalization, changing demographics and aspirations, first-time buyers, and availability of financing options are acting as growth drivers, according to a report by OLX Autos and rating agency Crisil.

"MyCars" is a new-age startup laying foundations in the setting up a car resell domain and they are setting up a team of ML experts to make predictive models determine the price of second-hand cars to optimize their revenue, you have joined as a new Data Scientist and your role is to create a model to determine the selling price of a used car.

## Objective:

1. Provide the best-performing model to determine the price of the used car.
2. Providing the most important features which determine the price.

## Data Description

The data provided consists of the following Data Dictionary - https://github.com/DsumitON/ML_With_Python/blob/main/cars.csv

1. Id: Unique ID assigned to a specific car.
2. year: Manufacture year of the car.
3. brand: Brand of the car.
4. full_model_name: Model name includes other details such as engine capacity,transmission, etc., basically a detailed model name.
5. model_name: Just the model name of the car.
6. price: Sell price of the 2nd ownership car.
7. distance_travelled(km): Distance traveled by car.
8. fuel_type : Fuel engine type .
9. city: City where the car is registered.
10. car_age: Age of the car.
